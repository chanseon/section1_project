## **Genre** (nan값30개는 지우는 걸로)

```python
check = df[df.Genre.isnull() == True]
check.sort_values(by=['Name'])
```

```python
df_sub = df.sort_values(by=['Name']).reset_index(drop=True) #Genre가 Nan인 애들을 채우기위해 이름정렬을 한다 -> Nan위에 바로 같은 애들이 있을 수 있다고 생각했는데 없는 경우도 있어서 다른 방안 선택
```

```python
genre_none = df_sub[df_sub.Genre.isnull()]['Name'] #Nan값을 갖는 애들의 이름을 추출, 그리고 각각 일치하는 게임이 있는지 확인
genre_none = list(genre_none)
```

```python
check_dic = {}
for i in range(len(genre_none)):
  memo = []
  for j in range(len(df_sub)):
    if genre_none[i] == df_sub['Name'][j]:
      memo.append(j)
      check_dic[genre_none[i]] = memo
```

```python
check_dic # genre 에서 None값의 제목과 같은 게임들의 index 값 추출 -> 한개만 있는 애들 외에는 중복되는 애들이 있으므로 해당 인덱스에는 같은 이름을 넣어준다.
```

```python
genre_none_duplicated=[] #해당 이름들이 genre가 Nan임에도 다른 플렛폼으로 존재하는 애들을 찾아서 genre 넣을 준비
for i in range(len(genre_none)):
    if len(check_dic.get(genre_none[i])) != 1:
      genre_none_duplicated.append(genre_none[i])
```

```python
genre_duplicated_counts = len(genre_none) - len(genre_none_duplicated)
genre_duplicated_counts #중복되지 않은 애들은 30개, 중복된 애들이 있어서 유추할 수 있는 애가 20개.
```

```python
for i in range(len(genre_none)):
    if len(check_dic.get(genre_none[i])) != 1:
      for j in range(len(check_dic.get(genre_none[i]))):
        if type(df_sub.Genre.loc[check_dic.get(genre_none[i])[j]]) == float:
          df_sub.Genre.loc[check_dic.get(genre_none[i])[j]] = df_sub.Genre.loc[check_dic.get(genre_none[i])[j-1]]
#중복된 애들 대체함   
```

```python
df_sub.Genre.isnull().sum() #중복되지 않아서 genre를 알 수 없는 애들 30개 확인 .. 지울까? 우선 NAN으로 수정
df_sub.Genre = df_sub.Genre.dropna()
```
